#Artificial Neural Network

##Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
tf.__version__
```

## Data Preprocessing

### Importing dataset

```python
df = pd.read_csv("/content/drive/MyDrive/data/Churn_Modelling.csv")
```

```python
df
```

```python
X = df.drop(["RowNumber","CustomerId","Surname","Exited"],axis=1)
```

```python
y = df["Exited"]
```

```python
X
```

```python
y
```

### Encoding categorical data

#### Label encoding the Gender column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["Gender"] = le.fit_transform(X["Gender"])
```

```python
X
```

```python
X["Geography"].unique()
```

```python
X[["France","Germany","Spain"]]= pd.get_dummies(X["Geography"])
```

```python
X = X.drop("Geography",axis=1)
```

### Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Building the ANN

###Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

### Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

## Training the ANN

### Compiling the ANN

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

### Training the ANN on the training set

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

## Making the predictions and evaluating the model

```python
print(ann.predict(sc.transform([[600,1,40,3,60000,2,1,1,50000,1,0,0]]))>0.5)
```

## Predicting the test set results

```python
y_preds = ann.predict(X_test)
```

```python
y_preds = (y_preds > 0.5)
```

```python
y_preds
```

```python
result = pd.DataFrame(y_test)
```

```python
result["prediction"]=y_preds
```

```python
result
```

## Making confusion matrix and accuracy

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```
